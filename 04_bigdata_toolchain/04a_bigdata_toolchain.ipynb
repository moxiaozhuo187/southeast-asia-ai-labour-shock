```python
#1.prepare big data input files
from google.colab import drive
drive.mount("/content/drive")
import pandas as pd
import csv
from pathlib import Path
drive_dir=Path("/content/drive/MyDrive/STQD6324_Labour_Shock")
out_dir=drive_dir/"outputs"
bigdata_dir=drive_dir/"bigdata-input"

bigdata_dir.mkdir(exist_ok=True)
core_df=pd.read_csv(out_dir/"worldbank_core_analysis_cleaned_2015_2024.csv")
index_df=pd.read_csv(out_dir/"country_year_evi_dri_index_2015_2024.csv")
cluster_df=pd.read_csv(out_dir/"sea_kmeans_cluster_2019_2024.csv")
print("core data shape:",core_df.shape)
print("index data shape:",index_df.shape)
print("cluster data shape:",cluster_df.shape)
```

```python
#2.create simple long table for Hive and Pig
tool_core=core_df[["country_code","country_name","is_benchmark",
    "indicator_code","indicator_group","year","value_cleaned"]].copy()

tool_core=tool_core.rename(columns={"value_cleaned":"value"})
tool_core_path=bigdata_dir/"tool_core_long_2015_2024.csv"
tool_core_noheader_path=bigdata_dir/"tool_core_long_2015_2024_noheader.csv"
tool_core.to_csv(tool_core_path,index=False,encoding="utf-8",quoting=csv.QUOTE_MINIMAL)
tool_core.to_csv(tool_core_noheader_path,index=False,header=False,encoding="utf-8",quoting=csv.QUOTE_MINIMAL)
print("saved:",tool_core_path)
print("saved:",tool_core_noheader_path)
print("tool_core shape:",tool_core.shape)
display(tool_core.head())
```

```python
#3.create simple index table for Hive and Spark
tool_index=index_df[["country_code","country_name","is_benchmark","year",
    "evi_main","evi_robust_010","evi_robust_020",
    "dri","risk_gap","risk_gap_robust_010","risk_gap_robust_020"]].copy()

tool_index_path=bigdata_dir/"tool_index_2015_2024.csv"
tool_index_noheader_path=bigdata_dir/"tool_index_2015_2024_noheader.csv"
tool_index.to_csv(tool_index_path,index=False,encoding="utf-8",quoting=csv.QUOTE_MINIMAL)
tool_index.to_csv(tool_index_noheader_path,index=False,header=False,encoding="utf-8",quoting=csv.QUOTE_MINIMAL)
print("saved:",tool_index_path)
print("saved:",tool_index_noheader_path)
print("tool_index shape:",tool_index.shape)
display(tool_index.head())
```

```python
#4.create simple kmeans table
tool_cluster=cluster_df[["country_code","country_name","evi_main_avg",
    "dri_avg","risk_gap_avg","cluster","cluster_label"]].copy()

tool_cluster_path=bigdata_dir/"tool_cluster_2019_2024.csv"
tool_cluster_noheader_path=bigdata_dir/"tool_cluster_2019_2024_noheader.csv"
tool_cluster.to_csv(tool_cluster_path,index=False,encoding="utf-8",quoting=csv.QUOTE_MINIMAL)
tool_cluster.to_csv(tool_cluster_noheader_path,index=False,header=False,encoding="utf-8",quoting=csv.QUOTE_MINIMAL)
print("saved:",tool_cluster_path)
print("saved:",tool_cluster_noheader_path)
print("tool_cluster shape:",tool_cluster.shape)
display(tool_cluster)
```

```python
#5.check comma risk
check_cols=["country_code","country_name","is_benchmark","indicator_code","indicator_group"]
for c in check_cols:
    comma_n=tool_core[c].astype(str).str.contains(",",regex=False).sum()
    print(c,"comma count:",comma_n)
print("tool_core rows:",tool_core.shape[0])
print("tool_index rows:",tool_index.shape[0])
print("tool_cluster rows:",tool_cluster.shape[0])
```

```python
#6.create bigdata input readme
readme_txt="""
Big Data Input Files

This folder stores the simplified CSV files used in the big data part of this project.

The original cleaned files are kept in the outputs folder. For Hive, Pig and Spark,
I created smaller and cleaner versions here so that the tools can read the data more easily.

Files in this folder:
1. tool_core_long_2015_2024.csv
Core indicator data in long format, with header.

2. tool_core_long_2015_2024_noheader.csv
Same as above, but without header. This version is mainly used for Hive and Pig.

3. tool_index_2015_2024.csv
Country-year EVI, DRI and risk gap results, with header.

4. tool_index_2015_2024_noheader.csv
Same as above, but without header. This version is mainly used for Hive.

5. tool_cluster_2019_2024.csv
SEA-only K-means clustering result, with header.

6. tool_cluster_2019_2024_noheader.csv
Same as above, but without header.

Why these files are simplified:
Some text fields in the original dataset, such as indicator_name, contain commas.
This may cause parsing problems when Hive or Pig reads the file using comma as the separator.
Therefore, these big data input files keep only the fields needed for Hive, Pig,
Spark and optional HBase tasks.
"""

readme_path=bigdata_dir/"README_bigdata_input.txt"
with open(readme_path,"w",encoding="utf-8") as f:
    f.write(readme_txt)
print("saved:",readme_path)
```
